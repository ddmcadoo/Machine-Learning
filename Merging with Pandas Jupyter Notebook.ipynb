```python
# import data from Excel 
fileName = 'Regional_Sales.xlsx' #ensure file is in same location as notebook or add path

xls = pd.ExcelFile(fileName)

# load individual sheets / we'll explain the data as we go along
eastern_region = pd.read_excel(xls, 'eastern_region')
western_region = pd.read_excel(xls, 'western_region')
sales_team = pd.read_excel(xls, 'sales_team')
store_name = pd.read_excel(xls, 'store_name', index_col='StoreID')
store_info = pd.read_excel(xls, 'store_info',index_col='StoreID')
product = pd.read_excel(xls, 'product')
```

## Concat
- Concatenate pandas objects along a particular axis.
- Default axis is 0

<a href="https://pandas.pydata.org/docs/reference/api/pandas.concat.html">pandas.concat</a>

```python
# using concat to join rows
sales = pd.concat([eastern_region, western_region])
sales 
```

```python
# use ignore_index to reset index when using concat
sales = pd.concat([eastern_region, western_region], ignore_index=True)
sales
```

## Join
- Join columns of another DataFrame
- Default is same as concat with axis of 1

<a href="https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.join.html">pandas.DataFrame.join</a>

```python
store_name
```

```python
store_info
```

```python
# use join to join columns
store = store_name.join(store_info)
store
```

```python
# notice that join is the same as concat with axis=1
store = pd.concat([store_name, store_info], axis=1)
store
```

## Merge
- Uses database-style joins

<a href="https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html">pandas.DataFrame.merge</a>

<img src='merge_options.png'>

```python
sales
```

```python
sales_team
```

```python
# default is "inner merge"
sales_temp = sales.merge(sales_team)
sales_temp
```

```python
# now let's look at the product table
product
```

```python
# notice that using the default inner merge in this case ends up with less rows
sales_temp = sales.merge(product)
sales_temp
```

```python
# left merge keeps all rows of sales DataFrame
sales_temp = sales.merge(product, how='left')
sales_temp
```

```python
# right merge keeps all rows of product DataFrame
sales_temp = sales.merge(product, how='right')
sales_temp
```

```python
# outer merge keeps all rows of both DataFrames
sales_temp = sales.merge(product, how='outer')
sales_temp
```

```python
# you can merge on  columns by passing a list of columns
sales_temp = sales.merge(store, on='StoreID') 
sales_temp
```
